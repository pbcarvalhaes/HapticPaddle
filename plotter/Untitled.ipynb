```python
import pandas as pd

dados = pd.read_csv("data.txt")

dados.head()
```

```python
plt.plot(dados.Time/1000, dados.Signal)
plt.ylabel("leitura do sensor")
plt.xlabel("Tempo(s)")
plt.grid()
plt.show()
plt.plot(dados.Time, [((x-8388607)/8388607)*80 for x in dados.Signal])
plt.ylabel("leitura do sensor estimado (mV)")
plt.xlabel("Tempo aproximado (ms)")
plt.grid()
plt.show()
```

```python
plt.plot(dados.Time, dados.Signal)
plt.ylabel("leitura do sensor")
plt.xlabel("Tempo (ms)")
plt.axis([34000, 44000, -400, 2600])
plt.grid()
plt.show()
plt.plot(dados.Time, [((x-8388607)/8388607)*80 for x in dados.Signal])
plt.ylabel("leitura do sensor estimado(mv)")
plt.xlabel("Tempo aproximado (ms)")
plt.axis([34000, 59000, -0.005, 0.024])
plt.grid()
plt.show()
```

```python
plt.plot(dados.Time, dados.Signal)
plt.ylabel("leitura do sensor")
plt.xlabel("Tempo aproximado (ms)")
plt.show()
```

```python
stable = dados[dados.Time > 30000]
stable["mv"] = stable["Signal"].apply(lambda x:((x-8388607)/8388607)*80)

print(stable)
print("Desvio padrao: \n",stable.std())
print("Media: \n",stable.mean())
```
